```python
# Filter the DataFrame to include only rows with 'XSS' or 'Benign' in the Attack column
filtered_df = df[df['Attack'].isin(['xss', 'Benign'])]

# Display the filtered DataFrame
filtered_df
```

```python

# Check the count of each class in the filtered DataFrame
class_counts = filtered_df['Attack'].value_counts()

# For classes with fewer than 50,000 samples, duplicate rows to reach 50,000
xss_df = filtered_df[filtered_df['Attack'] == 'xss']
benign_df = filtered_df[filtered_df['Attack'] == 'Benign']

if len(xss_df) < 50000:
    xss_df = xss_df.sample(n=50000, replace=True, random_state=42)  # Oversample with replacement
else:
    xss_df = xss_df.sample(n=50000, random_state=42)

if len(benign_df) < 50000:
    benign_df = benign_df.sample(n=50000, replace=True, random_state=42)  # Oversample with replacement
else:
    benign_df = benign_df.sample(n=50000, random_state=42)

# Combine the datasets
balanced_sampled_df = pd.concat([xss_df, benign_df])

# Shuffle the dataset
balanced_sampled_df = balanced_sampled_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Display the result
balanced_sampled_df
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Step 1: Splitting the Data
X = balanced_sampled_df.drop(columns=['Attack'])  # Drop target column
y = balanced_sampled_df['Attack'] # Encode target as 1 for XSS, 0 for Benign

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Step 2: Scaling the Features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 3: Train and Evaluate Models

# Helper function to train and evaluate models
def train_and_evaluate_model(model, model_name):
    # Train the model
    model.fit(X_train_scaled, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test_scaled)
    
    # Evaluate
    print(f"Classification Report for {model_name}:\n")
    print(classification_report(y_test, y_pred,digits=4))
    
    # Plot confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Benign', 'XSS'])
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f"Confusion Matrix: {model_name}")
    plt.show()

# Random Forest
rf_model = RandomForestClassifier(random_state=42)
train_and_evaluate_model(rf_model, "Random Forest")

# Gradient Boosting
gb_model = GradientBoostingClassifier(random_state=42)
train_and_evaluate_model(gb_model, "Gradient Boosting")

# XGBoost
xgb_model = XGBClassifier(random_state=42)
train_and_evaluate_model(xgb_model, "XGBoost")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier

# Function to perform grid search for hyperparameter tuning
def tune_and_evaluate_model(model, param_grid, model_name):
    # Perform grid search
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, scoring='f1', verbose=2, n_jobs=-1)
    grid_search.fit(X_train_scaled, y_train)
    
    # Best model after tuning
    best_model = grid_search.best_estimator_
    print(f"\nBest Parameters for {model_name}: {grid_search.best_params_}")
    
    # Evaluate the tuned model
    y_pred = best_model.predict(X_test_scaled)
    print(f"\nClassification Report for {model_name} (Tuned):\n")
    print(classification_report(y_test, y_pred, digits=4))  # 4 decimal places
    
    # Plot confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Benign', 'XSS'])
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f"Confusion Matrix: {model_name} (Tuned)")
    plt.show()

# Random Forest Tuning
rf_param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
tune_and_evaluate_model(RandomForestClassifier(random_state=42), rf_param_grid, "Random Forest")

# Gradient Boosting Tuning
gb_param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'min_samples_split': [2, 5, 10]
}
tune_and_evaluate_model(GradientBoostingClassifier(random_state=42), gb_param_grid, "Gradient Boosting")

# XGBoost Tuning
xgb_param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'colsample_bytree': [0.8, 1.0]
}
tune_and_evaluate_model(XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42), xgb_param_grid, "XGBoost")
```

```python
from sklearn.model_selection import cross_val_score
# Function to calculate cross-validation accuracy
def cross_validation_accuracy(model, model_name, X, y):
    scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')  # 5-fold CV
    print(f"{model_name} Cross-Validation Accuracy: {scores.mean():.4f} ± {scores.std():.4f}")
    return scores.mean(), scores.std()

# Default Models
rf_default = RandomForestClassifier(random_state=42)
gb_default = GradientBoostingClassifier(random_state=42)
xgb_default = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

print("\nDefault Models:")
rf_default_acc, rf_default_std = cross_validation_accuracy(rf_default, "Random Forest (Default)", X_train_scaled, y_train)
gb_default_acc, gb_default_std = cross_validation_accuracy(gb_default, "Gradient Boosting (Default)", X_train_scaled, y_train)
xgb_default_acc, xgb_default_std = cross_validation_accuracy(xgb_default, "XGBoost (Default)", X_train_scaled, y_train)

# Tuned Models
rf_tuned = RandomForestClassifier(
    n_estimators=200,
    max_depth=20,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42
)
gb_tuned = GradientBoostingClassifier(
    n_estimators=200,
    learning_rate=0.2,
    min_samples_split=10,
    random_state=42
)
xgb_tuned = XGBClassifier(
    n_estimators=200,
    learning_rate=0.2,
    colsample_bytree=0.8,
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)

print("\nTuned Models:")
rf_tuned_acc, rf_tuned_std = cross_validation_accuracy(rf_tuned, "Random Forest (Tuned)", X_train_scaled, y_train)
gb_tuned_acc, gb_tuned_std = cross_validation_accuracy(gb_tuned, "Gradient Boosting (Tuned)", X_train_scaled, y_train)
xgb_tuned_acc, xgb_tuned_std = cross_validation_accuracy(xgb_tuned, "XGBoost (Tuned)", X_train_scaled, y_train)
```
